```python
from main import KMeans
import numpy as np 
import matplotlib.pyplot as plt 

%matplotlib inline 
```

# 1. Implement image segmetation 

```python
def image_segment(img, K: int, dim: tuple, save_img=True, img_name='seg_img'): 
    """This function performs image segmentation on a given RGB(A) input image.
    
    Parameters: 
    - img: input image of shape (h, w, c) for height, width and channels. 
    - K: number of clusters for KMeans algorithm.
    - dim: dimension of shape (w, h) of the figure showing both the original image and the segmented image.
    """
    
    h, w, c = img.shape          # extract height and width of input image 
    X = img.reshape(-1, c)       # reshape the input image to form the data matrix X 
    
    kmeans = KMeans(K=K)
    kmeans.fit(X)
    centers, labels = kmeans.centers[-1], kmeans.labels[-1]
    
    # create a segmented image 
    seg_img = np.zeros(X.shape)
    # replace each pixel with its cluster
    for i in range(X.shape[0]): 
        seg_img[i] = centers[labels[i]]
        
    if c == 3: 
        seg_img = seg_img.astype('uint8')     # convert to integer for plotting with RGB image 
        
    # plotting the original image and the segmented image for comparison 
    fig = plt.figure(figsize=dim)
    
    fig.add_subplot(1, 2, 1)
    plt.imshow(img)
    plt.title("Original image")
    plt.axis("off")
    
    fig.add_subplot(1, 2, 2)
    plt.imshow(seg_img.reshape(h, w, c))
    plt.title(f"Segmented image with K = {K}")
    plt.axis("off")
    
    if save_img: 
        plt.savefig(f'images/{img_name}.jpg')
    
    plt.show()
```
